## 1. Calcular o negativo das imagens.

```python
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image

paths = ['./lena_gray_512.tif', './cameraman.tif', './house.tif']
images = [np.array(Image.open(path)) for path in paths]

def invert_image(img: np.ndarray) -> np.ndarray:
    return 255 - img

def process(img, cb: callable):
    inverted_img = cb(img.copy())

    plt.figure(figsize=(10, 10))

    plt1 = plt.subplot(1, 2, 1)
    plt1.imshow(img, cmap='gray')
    plt1.set_title('Original')

    plt2 = plt.subplot(1, 2, 2)
    plt2.imshow(inverted_img, cmap='gray')
    plt2.set_title('Modified')

    plt.show()

def for_each_imgs(cb: callable):
    for img in images:
        process(img, cb)

for_each_imgs(invert_image)

```

## 2. Diminuir pela metade a intensidade dos pixels

```python
def half_intensity(img: np.ndarray) -> np.ndarray:
    # divide a imagem por 2 para diminuir a intensidade pela metade
    return img // 10

for_each_imgs(half_intensity)
```

## 3. Incluir 4 quadrados brancos 10 x 10 pixels em cada canto das imagens.

```python
square_size = 10
square_color = 255

def add_squares(img: np.ndarray) -> np.ndarray:
    # draw squares in lateral sides
    img[:square_size, :square_size] = square_color
    img[:square_size, -square_size:] = square_color
    img[-square_size:, :square_size] = square_color
    img[-square_size:, -square_size:] = square_color

    return img

for_each_imgs(add_squares)
```

## 4. Incluir 1 quadrado preto 15X15 no centro das imagens.

```python
square_size = 15
square_color = 0

def add_square_center(img: np.ndarray) -> np.ndarray:
    # draw squares in lateral center
    shape = img.shape
    center = (shape[0] // 2, shape[1] // 2)

    img[center[0] - square_size // 2:center[0] + square_size // 2,
    center[1] - square_size // 2:center[1] + square_size // 2] = square_color

    return img

for_each_imgs(add_square_center)
```
